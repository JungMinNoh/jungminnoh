# the Lorenz equations - scipy 로 풀기 14108334 노정민

$$
\begin{align*}
x'(t) &amp;= \sigma(y - x)\\ 
y'(t) &amp;= x(\rho - z) - y \\ 
z'(t) &amp;= x y - \beta z
\end{align*}
$$

```
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
from scipy.integrate import solve_ivp
import sympy
```

```
def f(t, xyz):
    x, y, z = xyz
    return [sigma * (y - x),
            x * (rho - z) - y,
            x * y - beta * z]
```

```
xyz0 = [1.0, 1.0, 1.0]
t = (0, 25)
```

```
rho, sigma, beta = 28, 8, 8/3.0

xyz1 = solve_ivp(f, t, xyz0)
print(xyz1)
```

```
rho, sigma, beta = rho, sigma, 0.6*beta

xyz2 = solve_ivp(f, t, xyz0)
print(xyz2)
```

```
rho, sigma, beta = rho, 2*sigma, 0.6*beta

xyz3 = solve_ivp(f, t, xyz0)
print(xyz3)
```

```
from mpl_toolkits.mplot3d.axes3d import Axes3D

fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(12, 3.5), subplot_kw={'projection': '3d'})

for ax, xyz, c in [(ax1, xyz1, 'r'), (ax2, xyz2, 'b'), (ax3, xyz3, 'g')]:
    ax.plot(xyz.y[0], xyz.y[1], xyz.y[2], c, alpha=0.5)
    ax.set_xlabel('x', fontsize=16)
    ax.set_ylabel('y', fontsize=16)
    ax.set_zlabel('z', fontsize=16)
    ax.set_xticks([-15, 0, 15])
    ax.set_yticks([-20, 0, 20])
    ax.set_zticks([0, 20, 40])
```
